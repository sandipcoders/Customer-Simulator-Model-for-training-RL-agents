```python
import pandas as pd
import numpy as np
df_new=pd.read_csv("state_new_1.csv")
df_new.drop(["CDP","number_of_product","number_of_coupon_redem"
             ,"tran_value","match","household_key","rolling_3_tran_value"], axis = 1, inplace = True)
df_new.head()
```

```python
data_sorted = df_new.sort_values(by=['rolling_number_of_product'])

# Calculate the 75th percentile value
percentile_75 = data_sorted['rolling_number_of_product'].quantile(0.75)

# Filter the dataset to include only the values less than or equal to the 75th percentile value
df_new = data_sorted[data_sorted['rolling_number_of_product'] <= percentile_75]

# Save the filtered dataset
df_new.to_csv('myfiltereddataset.csv', index=False)
from scipy import stats
df_new=df_new[(np.abs(stats.zscore(df_new)) < 3).all(axis=1)]
df_new.to_csv('myfiltereddataset.csv', index=False)
```

```python
mean_df=df_new.mean(axis=0)
df_new=df_new-mean_df
std_df=df_new.std(axis=0)
df_new=df_new/std_df
```
